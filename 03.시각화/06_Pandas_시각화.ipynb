```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

## Line Plot

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title('Pandas의 Plot Method 사용 예', y = 1.05)
plt.xlabel('일자')
plt.ylabel('Data')
plt.grid()
plt.show()
```

## Bar Chart

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)      # Series
plt.title("꽃받침의 길이 시각화", y=1.05)
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0,
                            color=plt.get_cmap('Greys')(np.linspace(0.8, 0.2, 20)))
plt.title("꽃받침의 길이 시각화", y=1.05)
plt.xlabel("꽃받침의 길이")
plt.ylabel("Data")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.ylim(0,7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('각 Species의 Feature별 평균', y=1.05)
plt.ylim(0,8)
plt.show()
```

```python
df2.T
```

```python
df2.T.plot.bar(rot=0)
plt.title('각 Feature의 Species별 평균', y=1.05)
plt.show()
```

## Pie Chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수 비율', y=1.05)
plt.axis('equal')
plt.show()
```

## Histogram

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 Histogram", y=1.05)
plt.xlabel("데이터 값")
plt.grid()
plt.show()
```

## Box Plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot", y=1.05)
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=3, h_pad=1)
plt.show()
```
